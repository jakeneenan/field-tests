```python
df_h = pd.read_excel('../data/data.xlsx', sheet_name='h_statements')

df_h.head()
```

```python
df_h
```

```python
df_h.columns = df_h.columns.str.lower()
df_h.columns = df_h.columns.str.replace(' ', '_')
```

```python
df_h[df_h.code == 'H371']
```

```python
df.chemical = df.chemical.str.strip()
```

```python
chemicals = df.chemical.unique()



```

```python
df_h[df_h.code == 'H300'].health_hazard_statement.values[0]
```

```python
dfs = []




for c in chemicals:
    statements = []
    classes = []
    h_codes = []

    temp_chemical = df[df.chemical == c]

    
    try:
        codes = temp_chemical.ghus_us.values[0].split(',')
        codes = [c.strip() for c in codes]
    except:
        print("bad codes", codes, c)
        

    if 'na' not in codes:

        for code in codes:
            if code.startswith('H3'):
                statement = df_h[df_h.code == code].health_hazard_statement.values[0]
                temp_class = df_h[df_h.code == code]["hazard_class_(ghs_chapter)"].values[0]
                statements.append(statement)
                classes.append(temp_class)
                h_codes.append(code)
        
        temp_df = pd.DataFrame({'chemical': c, 'statement': statements, 'class': classes, 'code': h_codes})
        dfs.append(temp_df)
    



        
    
```

```python
df_statements = pd.concat(dfs)

df_statements
```
